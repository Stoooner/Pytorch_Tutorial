```python
import torch
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print(torch.__version__)
print(torch.cuda.is_available())
```

## 读取数据

```python
data = pd.read_csv(filepath_or_buffer="../小型CSV数据/credit-a.csv", header=None)
# data.info()
data
```

```python
# https://blog.csdn.net/u014712482/article/details/85080864
X = data.iloc[:, :-1]
Y = data.iloc[:, -1].replace(-1, 0) # Y值将取值为-1的标签替换为0
Y.unique() # unique函数去除其中重复的元素
```

```python
# 将数值转换为tentor
X = torch.from_numpy(X.values).type(torch.float32)
X.shape
```

```python
# Y = torch.from_numpy(Y.values.reshape(X.shape[0], 1)).type(torch.float32)
Y = torch.from_numpy(Y.values.reshape(-1, 1)).type(torch.float32)
```

## 模型搭建

```python
# 在线性回归章节中使用的是torch.nn.Linear()是因为那个仅仅是单个层
# torch.nn.Sequential()可以将多个层顺序地联合在一起
# torch.nn.Sigmoid在我们的神经网络中使用时，我们应该将其看作是网络的一层，而不是简单的函数使用。
# https://blog.csdn.net/W1995S/article/details/115429322?utm_medium=distribute.pc_relevant.none-task-blog-baidujs_title-0&spm=1001.2101.3001.4242
model = torch.nn.Sequential(
    torch.nn.Linear(in_features=X.shape[-1], out_features=Y.shape[-1]),
    torch.nn.Sigmoid()
)
```

```python
# 打印网络结构的方法1：
model
# 打印网络结构的方法2：
# https://blog.csdn.net/weixin_43275702/article/details/107465155
```

```python
# https://www.5axxw.com/questions/content/bdker3
# device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
# model.to(device=device)
# 打印网络结构的方法1：
from torchsummary import summary
summary(model, input_size=(X.shape[0], X.shape[1]))
# https://blog.csdn.net/weixin_42118374/article/details/105102153
# https://www.cnblogs.com/ywheunji/p/12368667.html
```

```python
loss_func = torch.nn.BCELoss()
optimizer = torch.optim.Adam(params=model.parameters(), lr=0.0001)
```

## 训练参数调整

```python
batch_size = 16
batch_num = X.shape[0] // batch_size
epochs = 1000
```

## 模型训练

```python
for epoch in range(epochs): # 总共训练轮数
    print("epoch = ", epoch)
    for i in range(batch_num): # 每轮训练的batch数
        start_index = i * batch_size
        end_index = start_index + batch_size
        x, y = X[start_index: end_index], Y[start_index: end_index]
        y_pred = model(x)
        loss = loss_func(y_pred, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
