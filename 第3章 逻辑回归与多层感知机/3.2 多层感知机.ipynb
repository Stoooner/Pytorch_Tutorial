## 创建模型
- 不使用Sequential，而是使用自定义模型的方式进行模型搭建
- nn.Module: 继承自这个类
- __init__: 初始化所有的层组件
- forward： 定义模型的前向传播过程

```python
class Model(torch.nn.Module):
    def __init__(self):
        # 继承父类中所有的属性
        super().__init__()
        self.linear_1 = torch.nn.Linear(in_features=20, out_features=64)
        self.linear_2 = torch.nn.Linear(in_features=64, out_features=64)
        self.linear_3 = torch.nn.Linear(in_features=64, out_features=1)
        self.activation_func = torch.nn.ReLU()
        self.sigmoid = torch.nn.Sigmoid()

    def forward(self, input): # 覆盖重写父类中的该方法
        x = self.linear_1(input)
        x = self.activation_func(x)
        x = self.linear_2(x)
        # 还可以使用函数式api： 
        # import torch.nn.functional as F
        # x = F.relu(self.linear_2(x)) 
        # 这样的话就不用再向上面那样在__init__中初始化激活函数组件了，直接使用即可
        x = self.activation_func(x)
        x = self.linear_3(x)
        x = self.sigmoid(x)
        return x
```

```python
learning_rate = 0.001

def get_model():
    model = Model()
    optimizer = torch.optim.Adam(params=model.parameters(), lr=learning_rate)
    return model, optimizer

model, optimizer = get_model()

#=========================================
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device=device)
from torchsummary import summary
summary(model, input_size=(14999, 20))
```

```python
loss_func = torch.nn.BCELoss()
batch_size = 64
batch_num = X.shape[0] // batch_size
epochs = 100
```

```python
for epoch in range(epochs):
    for i in range(batch_num):
        start_index = i * batch_size
        end_index = start_index + batch_size
        x, y = X[start_index: end_index], Y[start_index: end_index]
        y_pred = model(x)
        loss = loss_func(y_pred, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    with torch.no_grad():
        print("epoch = ", epoch, "loss = ", loss_func(model(X), Y).data.item())
        
```

## 使用DataLoader进行重构

```python
# 训练集和测试集划分
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset
train_x, test_x, train_y, test_y = train_test_split(X_data, Y_data)
# 由于现在划分出来的train_x等都是numpy.ndarray格式，不能直接放入到TensorDataset，需要转换为tensor
train_x = torch.from_numpy(train_x).type(torch.float32)
train_y = torch.from_numpy(train_y).type(torch.float32)
test_x = torch.from_numpy(test_x).type(torch.float32)
test_y = torch.from_numpy(test_y).type(torch.float32)
# 使用DataLoader装载数据
train_dataset = TensorDataset(train_x, train_y)
traindata_loader = DataLoader(dataset=train_dataset, 
                              batch_size=batch_size,
                              shuffle=True)
test_dataset = TensorDataset(test_x, test_y)
testdata_loader = DataLoader(dataset=test_dataset, 
                             batch_size=batch_size)

# 定义计算正确率的函数
def accuracy(y_pred, y_true):
    y_pred = (y_pred > 0.5).type(torch.int32)
    acc = (y_pred == y_true).float().mean()
    return acc
```

```python
for epoch in range(epochs):
    for x, y in traindata_loader:
        y_pred = model(x)
        loss = loss_func(y_pred, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    with torch.no_grad():
        epoch_accuracy = accuracy(model(train_x), train_y)
        epoch_loss = loss_func(model(train_x), train_y).data.item()

        test_accuracy = accuracy(model(test_x), test_y)
        test_loss = loss_func(model(test_x), test_y).data.item()
        print("epoch = ", epoch, "loss = ", round(epoch_loss, 3), 
              "accuracy = ", epoch_accuracy.item())
```
